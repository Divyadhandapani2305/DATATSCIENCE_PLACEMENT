## SERIES ##

```python
import pandas as pd
a = pd.Series(['divi','arthi','mouni','makizh'])
print(a)
print(a.values)
print(a.index)
print(a.dtype)

#Accessing Elements
print(a[1])

#Access Elements using loc
print(a.loc[2])
print(a.iloc[2])

#Slicing
print(a[1:3])
print(a.loc[1:3])
print(a.iloc[1:3])
```

```python
import pandas as pd
arr = pd.Series([12,25,36,40,50,12],index = ['A','B','C','D','E','F'])
print(arr[['A','C','F']])
print(arr)
print(arr['E'])

#Vectorized Operations - No loop required
print(arr+5)
print(arr-6)
print(arr*5)
print(arr.loc['A'])
print(arr.iloc[1])

#Statistical Function
print(arr.sum())
print(arr.max())
print(arr.min())
print(arr.std())
print(arr.mean())
print(arr.median())
print(arr.var())

#Boolean Indexing
print(arr[arr>30])

#Count Values
print(len(arr))
# it return indivitual value count
print(arr.value_counts())

#String Operation
name = pd.Series(['kaviya','mithra','priya'])
print(name.str.upper())
print(name.str.len())

```

## DATA FRAME ##

```python
import pandas as pd

#Method 1 : From Dictionary
data = ({
    'name':['arthi','mouni','divi'],
    'roll_no':[4,18,7],
    'age' : [21,22,21]})
df = pd.DataFrame(data)
print(df)

#Method 2 : From series to DataFrame
s1 = pd.Series([1,2,3])
s2 = pd.Series([4.9,5.4,6.3])
s3 = pd.Series([True,False,True])
df= pd.DataFrame({'sno' : s1,
                 'cgpa' : s2,
                 'attendence' : s3})
print(df)

#List of list
data =[['dharshini',22,98],
    ['dharani',21,88],
    ['abi',20,77]]
df=pd.DataFrame(data,columns=["name","age","mark"])
print(df)

#List of Dictionary
data =[
    {"name" : "bhuvana","course": "cs","grade":"A"},
    {"name" : "vicky","course": "bca","grade":"A++"},
    {"name" : "ragu","course": "cs","grade":"A"}]
df = pd.DataFrame(data)
print(df)


data =[
    {"name" : "bhuvana","course": "cs","grade":"A"},
    {"name" : "vicky","course": "bca","grade":"A++"},
    {"name" : "ragu","course": "cs","grade":"A"}]
df = pd.DataFrame(data,index = ["std1","std2","std3"])
print(df)


#Adding one extra Columns
df['marks']=[99,77,98]
print(df)

#Adding one extra Row
std4=pd.DataFrame([{"name":"kathir","course": "cs","grade":"A+","marks" : 90}],index=["std4"])
df=pd.concat([df,std4])
print(df)

#Create Multiple Rows
students = pd.DataFrame([{"name":"saran","course": "ds","grade":"A+","marks" : 90},
                        {"name":"sarath","course": "aids","grade":"A","marks" : 99}],index=["std5","std6"])
df = pd.concat([df,students])
print(df)

#Column Access by Bracket Notation and Dot Notation
print(df['marks'])
print(df.course)
print(df[['grade','marks']])

#shape attribute
print(df.shape)

#Find Number Of Rows
print(df.shape[0])

#Find Number Of Columns
print(df.shape[1])

#Dtype Attribute
#Used to check Datatype of all columns
print(df.dtypes)

#Info
print(df.info())

print(df.describe())

#Select Specific Rows
print(df.loc["std3"])

#Multiple Rows
print(df.loc[["std1","std4"]])

print(df.iloc[1])

#Update Value
df.loc['std1','grade']='B++'
print(df)
df.loc['std1','name']='Surya'
print(df)

#Delete Column
df.drop('grade',axis=1,inplace=True)
print(df)

#Delete Row
df.drop('std6',axis=0,inplace=True)
print(df)

#Sort Values
df.sort_values('marks',inplace=True)
print(df)

#Decending
df.sort_values('marks',ascending = False, inplace=True)
print(df)

#Statistical Function for particular Column
print(df['marks'].max())
print(df['marks'].min())
print(df['marks'].mean())
print(df['marks'].median())
print(df['marks'].std())
print(df['marks'].mode())
print(df['marks'].var())
```
