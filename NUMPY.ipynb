#### INDEXING AND SLICING ####

```python
#1d array indexing
import numpy as np
arr = np.array([10,20,35,48,50,60,70,80])
print(arr[0])
print(arr[-1])


#2d array indexing
arr2 = np.array([[10,20,30],
                 [40,50,60],
                 [22,33,44]])
print(arr2[1][1])


# 1D array slicing
print(arr[2:7])
print(arr[-7:-1])
print(arr[:5])
print(arr[2:])
print(arr[::2])

# 2D array slicing
print(arr2[2])
print(arr2[:,2])


#boolean indexing
print(arr[arr>60])
print(arr[arr%2==0])

#fancy indexing
print(arr[[0,3,6]])

arr[2]=100
print(arr)
```

#### MATHEMATICAL OPERATIONS IN NUMPY ####

```python
#1D Scalar Broadcasting
import numpy as np
a = np.array([1,2,3])
print(a+10)

#2D Scalar Broadcasting
b = np.array([[1,2,3],
              [4,5,6]])
print(b+10)

#Row Broadcasting
#always no of columns must be same
print(b+a)
```

#### matrix operations ####

```python
import numpy as np
a = np.array([[1,2],
             [4,5]])
b = np.array([[3,4],
             [5,6]])
print(a*b)
print(np.dot(a,b))
print(a@b)
print(np.matmul(a,b))

#Array Transpose 
print(a.T) #Row becomes Column
```

#### UNIVERSAL MATHEMATICAL FUNCTION ####

```python
import numpy as np
a = np.array([4,9,16])
print(np.sqrt(a))

#Absolute value
b = np.array([-4,-9,-16])
print(np.abs(b))

#log values
print(np.log(a))

#exponential
print(np.exp(a))
```

```python
#Round function
import numpy as np
arr = np.array([1.234,5,674])
print(np.round(arr,2))
print(np.ceil(arr))
print(np.floor(arr))
```

#### STATISTICAL FUNCTION ####

```python
#statistical function
import numpy as np
a = np.array([10,20,30,40,50])
print(np.mean(a))
print(np.median(a))
print(np.std(a))
print(np.max(a))
print(np.min(a))
print(np.sum(a))
print(np.percentile(a,25))
print(np.var(a))
print(np.argmin(a))
print(np.argmax(a))

b = np.array([1,2,3,4,5])
c = np.array([2,4,6,8,10])
print(np.corrcoef(b,c))
```

#### AGGREGATE FUNCTION ####

```python
import numpy as np
a = np.array([[10,20,30],
              [40,50,60]])
print(np.sum(a))
print(np.sum(a,axis=0)) #Column wise
print(np.sum(a,axis=1)) #Row wise
```

```python
import numpy as np
scores = np.array([[45,67,89],
                 [60,70,80],
                 [90,40,80]])
print(scores)

print(np.mean(scores))

print(np.sum(scores,axis=0))
print(np.sum(scores,axis=1))


min_score = np.min(scores)
max_score = np.max(scores)

print(np.mean(scores,axis=0))

normalized_scores = (scores - min_score) / (max_score - min_score)
print(normalized_scores)
```
